```python
test_url='https://www2.telenet.be/residential/nl/producten/internet-mobiel/easy-internet---mobile-5gb/'
from selenium import webdriver
from selenium.webdriver.common.by import By

opt=webdriver.ChromeOptions()
opt.add_argument("--headless")
driver = webdriver.Chrome(options=opt)
driver.get(test_url)

cookie_accept = driver.find_element(By.XPATH,"//button[@ id='onetrust-accept-btn-handler']")
cookie_accept.click()

ul = driver.find_elements(By.XPATH,"//div[@ class='cmp cmp-grouping aem-GridColumn aem-GridColumn--default--12']//ul")
for li in ul:
    print(li.text)
```
